### .apply() method:

Let's suppose, I want to grab the last four digit of CC Number.. So we'll make a function and use .apply() method

```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(4375220550950)
```

```python
# Let's apply to fetch last 4 digit of credit card
df['CC Number'].apply(last_four)
```

###### Now lets suppose, I want to make a categorical feature how much big the total bill is..by number of $ sign

```python
def Rank(num):
    if num<10:
        return '$'
    elif num>=10 and num<20:
        return '$$'
    elif num>=20 and num<30:
        return '$$$'
    else:
        return '$$$$'
```

```python
df['Rank-based Bill'] = df['total_bill'].apply(Rank)
```

```python
df['Rank-based Bill']
```

### Multiple column operation: np.vectorize(object):

Let's suppose we want to measure the tip qulity: if this is more than 15% of the bill then it is generous..otherwise normal.

```python
def tip_qual(tip,bill):
    if tip/bill>=0.15:
        return 'Generous'
    else:
        return 'Normal'
```

```python
df['tip_qual'] = np.vectorize(tip_qual)(df['tip'],df['total_bill'])
df['tip_qual']
```
